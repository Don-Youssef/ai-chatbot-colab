#Importing

```python
import json
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.text import one_hot, Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, GlobalAveragePooling1D, Dense
import pickle as pkl
import numpy as np
```

#loading JSON data from a file into a Python dictionary

```python
with open('/content/data.json') as file:
  data = json.load(file) # you can also get dataset from kaggle
print(data)
```

#extracting patterns, labels, and responses from a nested dictionary structure into separate lists for training a chatbot or similar natural language processing model

```python
traning_examples = []
traning_labels = []
traning_responses = []
for example in data['data']:
  for pattern in example['patterns']:
    traning_examples.append(pattern)
    traning_labels.append(example['label'])
    traning_responses.append(example['responses'])
print(traning_examples)
print(traning_labels)
print(traning_responses)
```

#cleaning text data by converting sentences to lowercase

```python
def clean_text(sentences):
  cleaned_sentences = []
  for sentence in sentences:
    sentence = sentence.lower()
    cleaned_sentences.append(sentence)
  return cleaned_sentences
cleaned_traning_examples = clean_text(traning_examples)
print(cleaned_traning_examples)
```

#encoding the training labels using Label Encoding

```python
encoder = LabelEncoder()
traning_labels = encoder.fit_transform(traning_labels)
print(traning_labels)
pkl.dump(encoder, open('encoder.pkl', 'wb'))
```

#converting the training labels into a one-hot encoded format using to_categorical

```python
traning_labels = to_categorical(traning_labels)
print(traning_labels)
```

#Tokenizing and Sequencing Text Data

```python
tokenizer = Tokenizer(num_words=100)
tokenizer.fit_on_texts(cleaned_traning_examples)
sequences = tokenizer.texts_to_sequences(cleaned_traning_examples)
print(sequences)
pkl.dump(tokenizer, open('tokenizer.pkl', 'wb'))
```

#Padding of Text Sequences

```python
paded_sequences = pad_sequences(sequences, truncating='pre',  maxlen=20)
print(paded_sequences)
```

#Defining and summarizing a Keras Sequential model for text classification

```python
vocabulary_size = 100
embedding_dimension = 10
max_length = 20
model = Sequential()
model.add(Embedding(input_dim=vocabulary_size, output_dim=embedding_dimension))
model.add(GlobalAveragePooling1D())
model.add(Dense(units=10, activation='relu'))
model.add(Dense(units=9, activation='relu'))
model.add(Dense(units=8, activation='softmax'))
sample_input = np.random.randint(vocabulary_size, size=(1, max_length))
model(sample_input)
model.summary()
```

#Compiling and training a Keras model for text classification

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(np.array(paded_sequences), traning_labels, epochs=2000)
model.save("model.h5")
```

#Preprocessing Text Input for Model Prediction

```python
input = 'hello'
sequences_input = tokenizer.texts_to_sequences([input])
paded_sequences_input = pad_sequences(sequences_input, truncating='pre',  maxlen=20)
print(paded_sequences_input)
```

#Predicting Intent using the Trained Model

```python
print(model.predict(np.array(paded_sequences_input)))
```

#Mapping Model Prediction to Class Label

```python
inverse_label_input = encoder.inverse_transform(model.predict(np.array(paded_sequences_input)).argmax(axis=1).reshape(1))
print(inverse_label_input)
```
